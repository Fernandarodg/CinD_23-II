## DESCRIPTION
The database is from the World Bank, it is oriented towards development objective number 8: Promote sustained, inclusive and sustainable economic growth, full and productive employment and decent work for all. 🌳


> ## The variables being used are:
*   Year
*   Country
*   GDP per capita (current US$)
*   Total population
*   Labor force participation rate, total ages 15-64 (modeled ILO estimate)
*   Percentage of wage workers, total (modeled ILO estimate)
*   Own-account workers, total (% of total employment) (modeled ILO estimate)
*   Vulnerable employment, total (% of total employment) (modeled ILO estimate)
*   Unemployment, total (% of total labor force) (modeled ILO estimate)
*   Unemployment, total youth (% of total labor force ages 15-24) (modeled ILO estimate)
*   Labor force, total
*   Employment in agriculture (% of total employment) (modeled ILO estimate)
*   Employment in industry (% of total employment) (modeled ILO estimate)
*   Employment in services (% of total employment) (modeled ILO estimate)





























### Characteristics:

> The data of the variables are in percentage terms. The analysis is worked on per year and the period to be handled is 6 years (2014-2020), in 160 countries.
📓




## The purposes or goals 📌

* Evaluate the effectiveness of policies and programs
* Compare the economic and employment situation between countries and regions.
*Analyze and indicate economic sectors with the highest labor demand
