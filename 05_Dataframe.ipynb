# **3.Cleaning**

## Column names

```python
# Year to year
d.rename(columns = {'Year':'year'}, inplace = True)

# Total population to pop
d.rename(columns = {'Total population':'pop'}, inplace = True)

# GPD per capita to GDP
d.rename(columns = {'GDP per capita (current US$)':'GDP'}, inplace = True)

# Percentage of wage workers, total (modeled ILO estimate) to wage workers
d.rename(columns = {'Percentage of wage workers, total (modeled ILO estimate)':'wage workers'}, inplace = True)

# Own-account workers, total (% of total employment) (modeled ILO estimate)) to Own-account workers
d.rename(columns = {'Own-account workers, total (% of total employment) (modeled ILO estimate))':'Own-account workers'}, inplace = True)

# Vulnerable employment, total (% of total employment) (modeled ILO estimate) to  Vulnerable employment
d.rename(columns = {'Vulnerable employment, total (% of total employment) (modeled ILO estimate)':'Vulnerable employment'}, inplace = True)

d.head()
```
