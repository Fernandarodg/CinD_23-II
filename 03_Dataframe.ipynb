# **3.Cleaning**

## Column names

```python
# Country Name to Country
d.rename(columns = {'Country Name':'Country'}, inplace = True)

# Country Code to Code
d.rename(columns = {'Country Code':'Code'}, inplace = True)

# GDP Per Capita to GDPPC
d.rename(columns = {'GDP Per Capita':'GDPPC'}, inplace = True)

# Education Expenditure (% GDP) to EducationEx
d.rename(columns = {'Education Expenditure (% GDP)':'EducationEx'}, inplace = True)

# Health Expenditure (% GDP) to HealthEx
d.rename(columns = {'Health Expenditure (% GDP)':'HealthEx'}, inplace = True)

d.head()
```

# **4. Indexing**

```python
# Duplicate the dataframe
d2 = d
```

```python
# Duplicate year as string / how to create a column
d2['Year_str'] = d2['Year'].astype(str)
d2.info()
```

```python
# Indexing
idx = pd.MultiIndex.from_product([d2['Country'].unique(),range(d2.Year.min(),d2.Year.max()+1)],
                                 names = ['Country', 'Year'])

## Attach the index
d2 = d2.set_index(['Country', 'Year'])

d2.head()
```

# **6. Descriptive Stadistics**

```python
# Basic
d2 [['GDPPC','EducationEx','HealthEx', 'R&D']].describe().applymap(lambda x: '{:.1f}'.format(x))
```

ANALYSIS: There is a great deal of variation in GDPPC, EducationEx, HealthEx, and R&D expenditure across countries.
The median values for each of these variables are higher than the mean values, indicating that there are a few countries with very high values that are skewing the mean.
The high standard deviations for each of these variables also indicate that there is a great deal of variation within each country. This means that there are some people within each country who are much wealthier, healthier, and more educated than others.
The high median R&D expenditure is particularly noteworthy. This suggests that many countries are investing heavily in research and development, which could lead to new technologies and innovations that improve the lives of people around the world.
This information can be used to make a number of inferences about the state of the world. For example, the high variation in GDPPC suggests that there is a great deal of inequality between countries. The high median R&D expenditure suggests that many countries are investing in the future, which is a positive sign. However, the high variation in EducationEx and HealthEx suggests that there are still many people in the world who do not have access to quality education and healthcare.

## **7. Basic Graph**

```python
# Scatter matrix
fig1 = px.scatter_matrix(d2, dimensions = ['GDPPC','EducationEx', 'HealthEx', 'R&D',], color = 'Year_str')
fig1.show()
```

```python
# Scatter matrix
fig1 = px.scatter_matrix(d2, dimensions = ['GDPPC','EducationEx', 'HealthEx', 'R&D',], color = 'Year_str')
fig1.update_traces(diagonal_visible = False)
fig1.show()
```

```python
# Scatter matrix
fig1 = px.scatter_matrix(d2, dimensions = ['GDPPC','EducationEx', 'HealthEx', 'R&D',], color = 'Year_str')
fig1.update_traces(diagonal_visible = False)
fig1.update_traces(showupperhalf = False)
fig1.show()
```

ANALYSIS:The diagonal of the plot shows the distribution of each variable individually. The other parts of the plot show the relationship between each pair of variables. For example, the upper left corner of the plot shows the relationship between GDP per capita and education expenditure.

The plot shows that there is a positive correlation between GDP per capita and all three of the other variables. This means that countries with higher GDP per capita tend to have higher education expenditure, health expenditure, and R&D expenditure.

The plot also shows that the relationship between GDP per capita and the other variables has strengthened over time. This is evident from the fact that the red dots (representing the later years) are more concentrated in the upper right corner of the plot than the blue dots (representing the earlier years).

Overall, the plot suggests that there is a strong relationship between GDP per capita and investment in education, health, and R&D. This relationship has strengthened over time, suggesting that countries are increasingly recognizing the importance of investing in these areas.

The correlation between GDP per capita and education expenditure is the strongest, followed by the correlation between GDP per capita and R&D expenditure. The correlation between GDP per capita and health expenditure is the weakest.
The relationship between GDP per capita and the other variables is not linear. There is a greater increase in education expenditure, health expenditure, and R&D expenditure as GDP per capita increases at higher levels.
There is a large variation in education expenditure, health expenditure, and R&D expenditure among countries with similar GDP per capita. This suggests that there are other factors, such as cultural values and government policies, that also influence investment in these areas.
